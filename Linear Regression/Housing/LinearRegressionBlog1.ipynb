## Non Linearity
***
** We can check if there is any non-linearity in the data by plotting the residual plot, fortunately seaborn already does this **

```python
def residplot(model,X,y,**kwargs):
    '''
    This function plots a regular residual plot
    Inputs:
        model = Trained sklearn model
        X: Predictors as a pandas dataframe
        y: Responses as a pandas dataframe
        kwargs: seaborn.residplot arguments
    Output:
        Matplotlib axes
    '''
    vals = {'Residuals': y-model.predict(X), 'Predicted Values': model.predict(X)}
    resdf = pd.DataFrame(vals)
    ax = sns.residplot(y='Residuals',x='Predicted Values', data=resdf, **kwargs)
    return ax
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(10,8)
residplot(lm,X_train,y_train, lowess = True)
```

** As you can see there is a U shape trend in the residual plot which suggests non-linearity in the data, I will now try to adress it by introducing this non-linearity into the model **

** I'm going to assume that because of the U shape only increasing first order interactions and second order predictors will take care of this problem, I'll use a pipeline and the PolynomialFeatures transformer to do this **

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
poly_lm = make_pipeline(PolynomialFeatures(2,include_bias=False),
                           LinearRegression())
```

```python
poly_lm.fit(X_train,y_train)
```

```python
predictions = poly_lm.predict(X_test)
for i,num in enumerate(predictions):
    if num<0:
        predictions[i] = 0
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('RMSLE:', np.sqrt(metrics.mean_squared_log_error(y_test, predictions)))
```

```python
inpipepf = poly_lm.named_steps['polynomialfeatures']
inpipelm = poly_lm.named_steps['linearregression']
```

```python
stats.summary(inpipelm,inpipepf.transform(X_train),y_train,inpipepf.get_feature_names(X.columns))
```

```python
fig,ax = plt.subplots()
fig.set_size_inches(10,8)
residplot(inpipelm,inpipepf.transform(X_train),y_train, lowess = True)
```

** At this point the non-linearity in the data has been greatly reduced however we have a considerably lower F-score and very bad test scores which means we're over fitting. This is most probably due to the increased number of features, so we will need to introduce a feature selection algorithm to reduce the number of features to a more managable amount **
***
** For this project I'll pick K number of features using f_regression feature selection, to pick the best K I'll plot the RMSLE of the model for all possible values of K **

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
RMSLEs = []
Ks = []
for k in range(1,36):
    poly_lm_freg = make_pipeline(PolynomialFeatures(2,include_bias=False), SelectKBest(f_regression,k),
                               LinearRegression())
    poly_lm_freg.fit(X_train,y_train)
    predictions = poly_lm_freg.predict(X_test)
    for i,num in enumerate(predictions):
        if num<0:
            predictions[i] = 0
    RMSLEs.append(np.sqrt(metrics.mean_squared_log_error(y_test, predictions)))
    Ks.append(k)
    
```

```python
fig, ax = plt.subplots()
ax.plot(Ks,RMSLEs)
fig.set_size_inches(15,5)
ax.set_xlabel('K')
ax.set_ylabel('RMSLE')
#assigning to suppress output
b = ax.set_xticks(range(36))
del b
```

## Outliers and High leverage observations
***
** First I'm going to examin the outliers of the training set and remove them, for that I'll need to plot the studentized residual plot against the predicted values and highlight the outliers **

```python
def rss(model,X,y):
    '''
    Function to calculate the residual sum of squared of the model
    Input:
        model: Trained linear regression model
        X: Training predictors as a pandas dataframe
        y: Training responses as a pandas dataframe
    Output:
        The RSS of the model
    '''
    res = y - model.predict(X)
    return sum(res**2)

def rse(model,X,y):
    '''
    Function to calculate the residual standard error of the model
    Input:
        model: Trained linear regression model
        X: Training predictors as a pandas dataframe
        y: Training responses as a pandas dataframe
    Output:
        The RSE of the model
    '''
    return np.sqrt(rss(model,X,y)/(X.shape[0]-X.shape[1]-1))

def studresidplot(model,X,y,threshold=3,ax=None, **kwargs):
    '''
    This function plots a studentized residual plot
    Inputs:
        model = Trained sklearn model
        X: Predictors as a pandas dataframe
        y: Responses as a pandas dataframe
        Threshold: Threshold to identify outliers
        kwargs: seaborn.residplot arguments
    Output:
        Boolean mask for outliers(False if outlier)
    '''
    vals = {'Studentized Residuals': (y-model.predict(X))/rse(model,X,y), 'Predicted Values': model.predict(X)}
    resdf = pd.DataFrame(vals)
    resdf['Outlier'] = resdf['Studentized Residuals'].apply(lambda x: 'Outlier' if abs(x)>= threshold else 'Normal')
    #ax = sns.lmplot(y='Studentized Residuals',x='Predicted Values', hue='Outlier' , data=resdf, fit_reg=False, **kwargs)
    ax = sns.scatterplot(y='Studentized Residuals',x='Predicted Values', hue='Outlier' , data=resdf, ax=ax, **kwargs)
    return resdf['Studentized Residuals'].apply(lambda x: False if abs(x)>= 3 else True)
```

```python
#Previous models need to be fitted and ran before running this for correct results
fig, ax = plt.subplots()
fig.set_size_inches(10,8)
mask_outliers = studresidplot(inpipelm,inpipesk.transform((inpipepf.transform(X_train))),y_train_scaled,ax=ax,threshold=2)
```

** To identify high leverage points it's necassary to compute the h statistics for each variable and one way to do that is computing the hat matrix **

```python
def hat_matrix(X):
    '''
    Function to compute the hat matrix based on the data matrix
    Inputs:
        X: data matrix as a np.ndarray matrix
    Outputs:
        H: hat matrix as a np.ndarray matrix
    '''
    ar = X
    art = X.transpose()
    H = np.matmul(np.matmul(ar,np.linalg.inv(np.matmul(art,ar))),art)
    return H

def h_stats(X):
    '''
    Function to compute the hat matrix based on the data matrix
    Inputs:
        X: data matrix as a np.ndarray matrix
    Outputs:
        h-statistics as a np.ndarray matrix
    '''
    H = hat_matrix(X)
    return np.diag(H)

def discard_high_leverage(X,y):
    hstats = h_stats(X)
    mask_leverage = hstats < (hstats.mean()*2)
    return X[mask_leverage],y[mask_leverage]
```
